# load our libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
# make some sample data so that we have something to plot
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50) #first value is our min, second value is our max, third value is how many value

# make our basic scatterplot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
# make a line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
# produce a histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
# Add axis labels and titles
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
# modify our labels and titlesʼ appearance by using font dictionaries

# make a dictionary with the properties we want for our labels
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
# add fontdict arguments to our previous lineplot
# move our lables with the loc argument
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
#  adjust the style and width of the line connecting our points
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,
y,
marker='*', 
color = 'indigo', 
linestyle = '--',   # NEW ADDITIONS
linewidth = 2)      
#fig.show()
```

```python
# modify colour using named colors or using hex codes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
 markersize = 12,              
# NEW ADDITIONS
 color = '#7425b9',
 linestyle = '--',
 linewidth = 2,
 markeredgecolor = '#fa9359',  # NEW ADDITIONS
 markerfacecolor = '#000000' ) # NEW ADDITIONS
```

```python
# add grid lines to our axis object using .grid()
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
 markersize = 12,              
# NEW ADDITIONS
 color = '#7425b9',
 linestyle = '--',
 linewidth = 2,
 markeredgecolor = '#fa9359',  # NEW ADDITIONS
 markerfacecolor = '#000000' ) # NEW ADDITIONS

ax.grid()
```

```python
#  modify the appearance (colour, width, linestyle) of our grid lines
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
 markersize = 12,              
# NEW ADDITIONS
 color = '#7425b9',
 linestyle = '--',
 linewidth = 2,
 markeredgecolor = '#fa9359',  # NEW ADDITIONS
 markerfacecolor = '#000000' ) # NEW ADDITIONS

#ax.grid(axis = 'y')
#ax.grid(axis = 'y', color = "blue", linewidth = 2, linestyle = '-.')

ax.grid(
    axis='y',                   # Add grid lines along the y-axis
    linestyle='-.',             # Dashed-dotted style
    color='#b9a25d',            # Golden grid line color
    linewidth=1.5,              # Grid line width
    alpha=0.8                   # Transparency level
)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
 markersize = 12,              
# NEW ADDITIONS
 color = '#7425b9',
 linestyle = '--',
 linewidth = 2,
 markeredgecolor = '#fa9359',  # NEW ADDITIONS
 markerfacecolor = '#000000' ) # NEW ADDITIONS

ax.grid(
    axis='x',                 # Display grid lines only on the x-axis
    linestyle='--',           # Dashed line style
    color='red',              # Red grid lines
    linewidth=2,              # Thicker grid lines for boldness
    alpha=1.0                 # Fully opaque
)
```
