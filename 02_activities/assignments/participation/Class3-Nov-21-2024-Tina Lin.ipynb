```python
# load our libraries and making some sample data

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# make a line plot with both of our y variables on the same axes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
# Add a label argument to each plot object
# Call legend and specify our legend location
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
```

```python
#  adding arguments to our legend object, we can change different elements of our legend

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
frameon = True, #add frame around the legend
fontsize = 12, #change font size
ncol = 2, #specify number of columns
shadow = True )
```

```python
# position the legend outside of our plot by
# Specifying legend location as before
# Specifying where we want to anchor that location on our figure using bbox_to_anchor

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
bbox_to_anchor =(1, 1))
```

```python
# position the legend outside of our plot by
# Specifying legend location as before
# Specifying where we want to anchor that location on our figure using bbox_to_anchor

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower left',
bbox_to_anchor =(1, 1))
```

```python
# position the legend outside of our plot by
# Specifying legend location as before
# Specifying where we want to anchor that location on our figure using bbox_to_anchor

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
bbox_to_anchor =(1, 0.5))
```

```python
# put our data into a scatter plot

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# use the text() function to specify coordinates and text
ax.text(10, 95, "This value is important!")
```

```python
# Modify our text annotation by adding arguments
# put our data into a scatter plot

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!",
ha='center',    
##alignment of text
color = 'red',  ##modify font colour
size = 20)      
```

```python
# Positioning text annotations
# make a simple, empty set of axes

fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData)           
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes)    
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure)
```

```python
# Annotating with arrows

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# Modifying arrows
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
arrowprops = dict(arrowstyle = "wedge",color = "hotpink"))
```

```python
# Removing tick marks and labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator()) # Use locator to remove ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter()) # Use formatter to remove only labels
```

```python
# Limiting tick marks

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# Tick mark intervals

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Rotating axis labels
# rotate axis labels and horizontal alignment, use xticks()

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right')
```

```python
# Modifying axis title fonts

font1 = {'family':'serif','color':'indigo'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha="right",
           fontsize = 12,
           font = "serif")
plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1)
```

```python
#use a style, just type the following before you make your figure

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

# define multiple axes and their relative positions
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))

ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
# Customizing our plots
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(9, 3))

font1 = {'family':'serif','size':10}
font2 = {'family':'serif','size':14}


ax1.scatter(x1,y1,marker='o',color='red')
ax2.bar(x2,y2,color='darkgreen')
```

```python
#Subplots without a grid arrangement

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

```

```python
#adding some very large x axis titles to our previous plot

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
 ['ax2', 'ax3']],
 figsize=(7, 4)) 
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18)
```
